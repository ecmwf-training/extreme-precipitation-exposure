# Explanation

In this section, you’ll find a clear description of how drought develops, how they can be measured, and strategies on how adaptation measures to reduce their impact. The explanation begins with describing the three **main drought types — meteorological, agricultural, and hydrological —** and how indices like SPI and SPEI help track them across different timescales.
You’ll also learn how drought risk is shaped by hazard, exposure, and vulnerability, and what practical adaptation options exist to address each of these dimensions. Together, these sections provide a structured foundation for understanding drought from early signals to long-term impacts — and what can be done to manage them.
