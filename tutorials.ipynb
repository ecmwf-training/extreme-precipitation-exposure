# Tutorials

Welcome to the tutorial collection!  

Here you'll find a series of hands-on guides that let you explore, analyze, and visualize drought (in conjuction with exposure and vulnerability) data — from quick introductions to more detailed workflows.

Under **Storylines**, you'll find risk-oriented tutorials (often using Central Greece as an example region):
- [Drought duration: Past, Present, Future changes](tutorials/changing_drought_duration.ipynb)
- [People experiencing prolonged droughts](tutorials/people_exposed_to_prolonged_droughts.ipynb)

Under **Explore Exposure Data**, you'll find tutorials that read and visualise exposure data sets: 
- [Understanding Population Data](tutorials/population.ipynb)
- [Understanding Agricultural Land Cover Changes](tutorials/agricultural_land.ipynb)

💡 **Tip:** Many of these tutorials can be run with data already stored in the GitHub repository. If you'd like to create similar datasets for a different region, check out the [how-to guides](how-to-guides.ipynb)!


**The jupyter notebooks in these tutorials can be run on free online platforms, such as Binder, Kaggle and Colab, or they can be accessed from GitHub. The links to run this notebook in these environments are provided here, but please note they are not supported by ECMWF.**
