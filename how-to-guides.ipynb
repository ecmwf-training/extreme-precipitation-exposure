# How-To Guides

Welcome to the how-to guides!

Here you'll find step-by-step instructions to help you download, reproduce and process drought and exposure data for your own regions. These guides cover everything from downloading data and defining custom regions to working with specific datasets.

You'll find a few basic guides that help you set up your analysis:
- [Define Your Region](how-to-guides/define_your_region.ipynb)
- [Download Data from the CDS](how-to-guides/download_data_from_cds.ipynb)

Under **Process Exposure Data**, you'll find guides for working with exposure datasets:
- [Population Data (GHS)](how-to-guides/population_ghs.ipynb)
- [Agricultural Land Cover (Satellite)](how-to-guides/agricultural_land_satellite.ipynb)
- [Agricultural Land Cover (CORINE)](how-to-guides/agricultural_land_corine.ipynb)

Under **Process Hazard Data**, you'll find guides for working with hazard datasets:
- [Drought Duration Data (ECDE)](how-to-guides/drought_duration_ecde.ipynb)

💡 **Tip:** Once you've created or downloaded your data, head over to the [tutorials](tutorials.ipynb) to explore and visualize your datasets in action!


**The jupyter notebooks in these tutorials can be run on free online platforms, such as Binder, Kaggle and Colab, or they can be accessed from GitHub. The links to run this notebook in these environments are provided here, but please note they are not supported by ECMWF.**
